```python
from sklearn.impute import SimpleImputer
```

```python
gender_column = datalar[['Gender']]
imputer = SimpleImputer(strategy='most_frequent')
imputed_gender = imputer.fit_transform(gender_column)
datalar['Gender'] = imputed_gender.reshape(-1)
```
