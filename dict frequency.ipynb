```python
#selecting more than one item orders
order_count=df['Order ID'].value_counts().to_dict()
order_count={v: k for v,k in order_count.items() if k>1}
```

```python
#finding together sold items by order id
together_items={}
for c in order_count.keys():
    list_of_products=df[df['Order ID']==c]['Product'].to_list()
    together_items[c]=list_of_products
    
```

```python
#finding how many times the product sold with other product or alone
product_frequency_dict=df['Product'].value_counts().to_dict()
```

```python
product_frequency_dict
```

```python
#finding how many times product sold together
product_list=list(df['Product'].unique())

new_prodyct_dict={}
for pr in product_list:
    for k, v in together_items.items():
        if pr in v:
            for val in v:
                if val!=pr:
                    new_key=f"{pr}+{val}"
                    if new_key not in new_prodyct_dict:
                        new_prodyct_dict[new_key]=1
                    else:
                        new_prodyct_dict[new_key]+=1
                        
import operator
new_prodyct_dict_sorted=dict(sorted(new_prodyct_dict.items(),key=operator.itemgetter(1),reverse=True))
                        
```

```python
new_prodyct_dict_sorted
```

```python
#divide to the overall count for finding probablity
final_dict={}
for npr in new_prodyct_dict_sorted.keys():
    first=npr.split('+')[0]
    final_dict[npr]=round(new_prodyct_dict_sorted[npr]/product_frequency_dict[first],2)
    
final_dict_sorted=dict(sorted(final_dict.items(),key=operator.itemgetter(1),reverse=True))
```
